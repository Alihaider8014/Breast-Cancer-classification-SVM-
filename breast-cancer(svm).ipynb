```python
cancer = load_breast_cancer()
```

```python
df=pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
df.tail()
```

```python
#visualizing the data
sns.pairplot(df, hue='target', vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
sns.countplot(df['target'])
```

```python
sns.scatterplot(x='mean area', y='mean smoothness', hue='target', data=df)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(),annot=True )
```

```python
X = df.drop(['target'], axis=1)
X.head()
```

```python
y= df['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model=SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
#evaluating the model
predict=svc_model.predict(X_test)
predict
```

```python
c=confusion_matrix(y_test,predict)
c
```

```python
sns.heatmap(c)
```

```python
#improving model
min_train = X_train.min()
```

```python
range_train = (X_train-min_train).max()
```

```python
X_train_scaled=(X_train - min_train)/range_train
```

```python
#before
sns.scatterplot(x=X_train['mean area'], y=X_train['mean smoothness'], hue=y_train)
```

```python
#after
sns.scatterplot(x=X_train_scaled['mean area'], y=X_train_scaled['mean smoothness'], hue=y_train)
```

```python
min_test=X_test.min()
range_test=(X_test - min_test).max()
X_test_scaled=(X_test - min_test)/range_test
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
predict=svc_model.predict(X_test_scaled)
```

```python
c=confusion_matrix(y_test, predict)
```

```python
sns.heatmap(c, annot=True)
```

```python
print(classification_report(y_test, predict))
```

```python
#improving the model_2
param_grid={'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gridd = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
```

```python
gridd.fit(X_train_scaled, y_train)
```

```python
gridd.best_params_
```

```python
predict=gridd.predict(X_test_scaled)
```

```python
sns.heatmap(c, annot=True)
```
